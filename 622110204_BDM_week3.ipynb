```python
# Ex 4. Randomly create a data matrix of 10 samples with 5 features. 
# Then, randomly create a label vector for 10 samples. 
# Finally, seperate samples into 2 groups
# Group 1 called train_data (label == 0)
# Group 2 called test_data (label == 1)
สร้างเมทริก สุ่มข้อมุล 10แถว5คอลัม
สร้างเวกเตอรที่ขนาดเท่ากับ10 1อัน
มีแค่ 0 1 เป็นหลัก
เปลี่ยนข้อมูล2กลุ่ม
กล่ม1 มีlebel=0 กลุ่ม2lebel=1 
```

```python
#Randomly create a data matrix of 10 samples with 5 features.
matrix = np.random.randint(10,size = (10,5))
matrix
```

```python
# Then, randomly create a label vector for 10 samples.
label_vector = np.random.randint(2,size = 10)
label_vector
```

```python
# Finally, seperate samples into 2 groups
# Group 1 called train_data (label == 0)
# Group 2 called test_data (label == 1)

train_data = matrix[np.where(label_vector == 0)]
train_data
```

```python
test_data = matrix[np.where(label_vector == 1)]
test_data
```
