# **3. Graph Type**

```python
import matplotlib.pyplot as plt

fig = plt.figure(1, figsize=(15,6))

# bar chart example
accounts = ['fabiyamada', 'jedigeiss', 'scipio', 'stoodkev']
steem = [1100, 500, 1200, 2300] #เอาค่าsteemมา+กันก่อน

plt.subplot(1,2,1)
plt.title('Bar chart')
plt.ylabel('Steem')
plt.bar(accounts, steem) #account แกนx steem แกน y

# pie chart example
labels = ['fabiyamada', 'jedigeiss', 'scipio', 'stoodkev']
perc = [steem[0]/sum(steem), # calculate percentages from the amount of steem
        steem[1]/sum(steem), 
        steem[2]/sum(steem), 
        steem[3]/sum(steem)]
#พายชาร์ตเป็น % เท่านั้น

explode = (0, 0, 0.5, 0)     # "explode" the 3rd slice ('Scipio') เอาอ้นที่3ระเบิดออกมาเพื่อเน้นข้อมูลให้ดูตัวนี้

plt.subplot(1,2,2)
plt.title('Pie chart')
plt.axis('equal')            # Make sure the pie is drawn as a circle, X and Y axes equally sized
plt.pie(perc, labels=labels, autopct='%1.2f%%', #กำหนดให้มี2ตำแหน่ง 1.2
        startangle=0, explode=explode)

plt.show()
```
